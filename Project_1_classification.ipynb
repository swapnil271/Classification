```python
x = dataset_train.iloc[:,:-1].values
pd.DataFrame(x)
```

```python
y = dataset_train.iloc[:,-1:].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Classification_Logistic_Regression_Algorithm

from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
```

```python
y_pred = logmodel.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
# Classification_K-Neighbors_Classifier_Algorithm

from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors=5, 
                                      metric='minkowski', p=2)
classifier_knn.fit(x_train, y_train)
```

```python
y_pred = classifier_knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
# Classification_Naive_Bayes_Algorithm

from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(x_train, y_train)
```

```python
y_pred=classifier_nb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
# Classification_SVC(Support Vector Machine)_Algorithm:-kernel='sigmoid'

from sklearn.svm import SVC
classifier_svm = SVC(kernel='sigmoid')
classifier_svm.fit(x_train,y_train)
```

```python
y_pred = classifier_svm.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Classification_SVC(Support Vector Machine)_Algorithm:-kernel='linear'

from sklearn.svm import SVC
classifier_svm = SVC(kernel='linear')
classifier_svm.fit(x_train,y_train)
```

```python
y_pred = classifier_svm.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Classification_SVC(Support Vector Machine)_Algorithm:-kernel='rbf'

from sklearn.svm import SVC
classifier_svm_rbf = SVC(kernel='rbf')
classifier_svm_rbf.fit(x_train,y_train)
```

```python
y_pred = classifier_svm_rbf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Classification_Decision_Tree_Classification_Algorithm

from sklearn.tree import DecisionTreeClassifier
classifier_dt = DecisionTreeClassifier(criterion='entropy')
classifier_dt.fit(x_train,y_train)
```

```python
y_pred = classifier_dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Classification_Random_Forest_Classification_Algorithm

from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier(n_estimators=3, criterion='entropy')
classifier_rf.fit(x_train,y_train)
```

```python
y_pred = classifier_rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
